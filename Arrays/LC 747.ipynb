```python
# Arrays: (LC 747)
# Problem Link: https://leetcode.com/problems/largest-number-at-least-twice-of-others

nums = list(map(int,input("Enter array of integers: ").split(',')))
nums1 = nums.copy()

largest = max(nums)
nums1.remove(largest)

for i in range(len(nums1)):
    if largest < nums1[i]*2:
        print(-1)
        break
else:
    print(nums.index(largest))
```
