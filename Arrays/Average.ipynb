```python
# Arrays: (LC 2455)
# Problem Link: https://leetcode.com/problems/average-value-of-even-numbers-that-are-divisible-by-three

nums = list(map(int,input("Enter space separated integers: ").split(',')))

res = []

for i in range(len(nums)):
    if nums[i]%2 == 0 and nums[i]%3 == 0:
        res.append(nums[i])

if not res:
    print(0)
else:
    avg = sum(res)//len(res)
    print(avg)
```
