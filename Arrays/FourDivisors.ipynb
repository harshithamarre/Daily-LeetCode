```python
# LC 1390
# Problem Link: https://leetcode.com/problems/four-divisors?envType=daily-question&envId=2026-01-04

import math

def divisors_of_number(nums):
    tsum = 0
    
    for n in nums:
        divisors = set()
        for i in range(1, int(math.isqrt(n)) + 1):
            if n % i == 0:
                divisors.add(i)
                divisors.add(n // i)
                if len(divisors) > 4:
                    break
    
        if len(divisors) == 4:
            tsum += sum(divisors)
            
    return tsum    

nums = list(map(int,input("Enter integer array: ").split(',')))
print(divisors_of_number(nums))
```
