```python
# Arrays: (LC 3512)
# Problem Link: https://leetcode.com/problems/minimum-operations-to-make-array-sum-divisible-by-k

nums = list(map(int,input("Enter array of integers: ").split(",")))
k = int(input("Enter any integer: "))

tsum = sum(nums)

if tsum % k == 0:
    print("0")
else:
    res = tsum % k
    print(res)
```
