```python
# LC 1351
# Problem Link: https://leetcode.com/problems/count-negative-numbers-in-a-sorted-matrix?envType=daily-question&envId=2025-12-28

def negative_numbers_count(grid):
    cnt = 0
    for row in grid:
        for val in row:
            if val < 0:
                cnt += 1
    return cnt

t1 = negative_numbers_count([[4,3,2,-1],[3,2,1,-1],[1,1,-1,-2],[-1,-1,-2,-3]])
print(t1)
t2 = negative_numbers_count([[3,2],[1,0]])
print(t2)
```
