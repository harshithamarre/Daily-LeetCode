```python
# Arrays: (LC 217)
# Problem Link: https://leetcode.com/problems/contains-duplicate

nums = list(map(int,input("Enter array of integers: ").split(",")))
unique = set(nums)

if len(unique) == len(nums):
    print("False")
else:
    print("True")

# ---- "Or" ----

uniq = []
duplicate = []

for i in range(len(nums)):
    if i not in uniq:
        uniq.append(nums[i])
    else:
        duplicate.append(nums[i])

if duplicate:
    print("True")
else:
    print("Talse")
```
